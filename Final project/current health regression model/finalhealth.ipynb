```python
df2 = pd.read_csv(r"D:\Final project\current health regression model\p1_train.csv")
df2
```

```python
df2.isnull().sum()
```

```python
# Renaming the names of column

df1.columns = ['col1','col2','target']
df2.columns = ['col1','col2','target']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error

x_train = df2[['col1','col2']]
y_train = df2['target']

x_test = df1[['col1','col2']]
y_test = df1['target']

models = [LinearRegression(),
          SVR()]
for model in models:
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    MSE = mean_squared_error(y_test,y_pred)
    MAE = mean_absolute_error(y_test,y_pred)

    print(f"********{type(model).__name__}********")
    print(f"Mean Squared Error (MSE): {MSE:.2f}")
    print(f"Mean Absolute Error (MAE): {MAE:.2f}")
```
